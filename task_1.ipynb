```python
# Importing all the necessary libraries 

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset using pandas 
df = pd.read_csv('/content/student_scores.csv')
```

```python
# Splitting the data into x train and y train  
X_train = df.iloc[:, :-1].values
y_train = df.iloc[:,-1].values
```

```python
# Creating model in Tensorflow
model = tf.keras.Sequential([
  tf.keras.layers.Dense(units=1, input_shape = [1])                            
 ])
```

```python
# Assigning Loss Function and Optimiser 
model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.SGD())
```

```python
# running the model 
history = model.fit(X_train, y_train, epochs=1000)
```

```python
# Predicting the value 
print(model.predict([9.25]))
```

```python
# Plotting the result 
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title('No.of hours Study vs Score in exams  (Training set TensorFlow)')
plt.xlabel('No. of Hours Studied')
plt.ylabel('Score')
plt.show()
```
